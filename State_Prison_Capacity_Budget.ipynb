## Capacity

```python
# Importing the dataset with the information
Prison_capacity=pd.read_excel("/content/drive/MyDrive/MDA-2022-Italy/Report/Dataset_preparation/Prison_Capacity_Stats_2020.xlsx")
Prison_capacity
```

```python
# Selecting the variables to compute the average capacity
df_avcap=Prison_capacity[["State","Lowest capacity","Highest capacity"  ]]
type(df_avcap['Lowest capacity'])
type(df_avcap['Highest capacity'])
```

```python
# Imputing the two missing values equals to 0.0
df_avcap.at[6,'Lowest capacity']=0.0
df_avcap.at[6,'Highest capacity']=0.0
df_avcap.at[34,'Lowest capacity']=0.0
df_avcap.at[34,'Highest capacity']=0.0
df_avcap
```

```python
# Creating the variable avg capacity
df_avcap['avg capacity']= float()
for i in df_avcap['State']:
  df_avcap['avg capacity']=(df_avcap['Lowest capacity']+ df_avcap['Highest capacity'])/2 
df_avcap
```

```python
# Removing the variable that we don't need anymore
del df_avcap['Lowest capacity']
del df_avcap['Highest capacity']
df_avcap
```

```python
# Changing missing values
df_avcap['avg capacity'].replace(0.00,np.NaN)
```

## Budget

```python
Prison_budget=pd.read_excel("/content/drive/MyDrive/MDA-2022-Italy/Report/Dataset_preparation/Prison_Budget.xlsx")
Prison_budget
```

```python
# Create the variable with the percentage of the budget given to prison
Prison_budget['Perc_Budget'] = float()
for i in Prison_budget['States']:
  Prison_budget['Perc_Budget'] = (Prison_budget['Prison_Budget']/Prison_budget['State_Budget'])*100

Prison_budget
```

```python
df_avcap['States']= Prison_budget['States']
```

```python
print(df_avcap)
```

## Merge the two Datasets together

```python
merged_total = pd.merge(df_avcap, Prison_budget , on=["States"])
del merged_total['State']
merged_total
```

```python
# Removing the useless variables
del merged_total['Prison_Budget']
del merged_total['State_Budget']
merged_total = merged_total.reindex(columns=['States','avg capacity','Perc_Budget'])
merged_total
```
